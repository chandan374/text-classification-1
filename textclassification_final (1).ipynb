# HERE ARE WE ARE MAKING THE CLASSFIER MANUALLY FROM SRATCH
....FIT FUNCTION...

```python


import numpy as np
def fit(x_train,y_train):
    result={}                     #DICTIONARY 
    class_values=set(y_train)     #DIFF_CLASSES
    
    for current_class in class_values: #...here we are traversing each class
        result[current_class]={}       #...now taking classes as a key building nested dictionary 
        result["total_data"]=len(y_train) #...saving the total rows values
        
        current_class_rows=(y_train==current_class) 
        x_train_current=x_train[current_class_rows]
        y_train_current=y_train[current_class_rows]
        
        num_feature=x_train.shape[1]               #...extracting all the features
        result[current_class]["total_count"]=len(y_train_current)  #this the TOTAL_COUNT column which saves size_of_each_class
        
        sum=0
        
        for j in range(1,num_feature+1):
            a=result[current_class][j-1]=(x_train_current.iloc[:,j-1]).sum() #here_we_are_counting_the_frequncy_of_each_WORD_in 
            sum=sum+a                                                        #particular_CLASS  
        
        result[current_class]["total_countfeat"]=sum                         #TOTAL_COUNT_OF_FEATURE putting in a new COLUMN
    
    return result                                                            #returning thr resulted dictionary
```

```python
dictionary=fit(x_train,y_train)     #calling the fiT_funtion
```

# THIS IS THE PROBABLITY FUNCTION
#this function we are finding the probablity 

```python

 
'''
    P(xj=xj/y=ci)={
                       COUNT[ci][word][file]  count_of_word_in_class_in_file 
                       /
                       count[c][total_words] count_of_total_wors_in_ci_class
                       }
                  * probablity(y=ci)
'''
def probablity(dictionary,x,current_class):
    
    output=np.log(dictionary[current_class]["total_count"])-np.log(dictionary["total_data"]) #prob(y==ci)
    
    x=list(x)              #listing the values_of_xth_row                            
    
    for j in range(len(x)):#trevesing the x row
        if x[j]==0:        #if the frequncy of the word in xth row 0 no IGNORE IT
            continue
        
        count_current_class_with_value_xj=dictionary[current_class][j]+1      #numerator && LAPLACE CORRECTION
        count_current_class_words=dictionary[current_class]["total_countfeat"]+len(dictionary[current_class].keys())-2#denominator
        current_xj_probablity=np.log(count_current_class_with_value_xj)-np.log(count_current_class_words)#finding_prob
        
        output=output+current_xj_probablity #summing up all the probablity
    
    return output                           #retuning probablity x row
        
        
```

# #THIS FUNCTION USED TO CHOOSE THE MAX PROBABLITY FITTING THE BEST CLASS

```python

def predictsinglepoint(dictionary,x):
    classes=dictionary.keys() #ALL the classes
    best_p=-1000              #initilising with minimum
    best_class=-1           #initialization
    first_run=True
    for current_class in classes:   #treversing each class
        if(current_class=="total_data"):  
            continue
        p_current_class=probablity(dictionary,x,current_class)    
        if(first_run or p_current_class>best_p):
            best_p=p_current_class
            best_class=current_class
            first_run=False
    return best_class           #returning the best suited class
```

# THIS IS THE PREDICT FUNTION
#here we are 
#calling the single_point_funtion 
#and passing the rows on_by_one

```python
def predict(dictionary,x_test):
    
    y_pred=[]
    num_feature=x_test.shape[0]
    
    for x in range(num_feature):
        x_class=predictsinglepoint(dictionary,x_test.iloc[x,:])
        y_pred.append(x_class)
    
    return y_pred    #getting the PREDICTION_OF_THE_TRAINIG_DATAFRAME
```
